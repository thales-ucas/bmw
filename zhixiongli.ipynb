```python
import json

with open('data1.json', 'r', encoding='utf-8') as file:
  data1 = json.load(file)

with open('data2.json', 'r', encoding='utf-8') as file:
  data2 = json.load(file)
with open('data3.json', 'r', encoding='utf-8') as file:
  data3 = json.load(file)

records = data1['records'] + data2['records'] + data3['records']
records
```

```python
len(records)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(records, columns=['articleTitle', 'publicationYear', 'publicationTitle', 'authors'])
df
```

```python
def get_author_str(authors):
  arr = []
  for author in authors:
    arr.append(author['preferredName'])
  return ', '.join(arr)

article = pd.DataFrame(columns=['Title', 'Year', 	'Publication',	'Affiliation of L',	'Co-Author',	'Editorial Board'])
article['Title'] =df['articleTitle']
article['Year'] = df['publicationYear']
article['Publication'] = df['publicationTitle']
article['Co-Author'] = df['authors'].apply(get_author_str)
article
```

```python
article.to_excel('article.xlsx')
```
